```python
from qiskit.quantum_info import Statevector
from qiskit import QuantumCircuit
circ = QuantumCircuit(3)
circ.h(0)
circ.cx(0, 1)
circ.cx(1, 2)
circ.draw('mpl')
state = Statevector.from_int(0,2**3)
state = state.evolve(circ)
state.draw('latex')
```
